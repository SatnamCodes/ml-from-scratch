## Linear Regression

#### Linear regression is essentially the "Hello World" of machine learning. It falls under supervised learning because the model learns from a labeled dataset—meaning it already knows the "answers" (targets) for the training data it’s given.At its core, the algorithm tries to find the best possible relationship between your input (independent variable) and your output (dependent variable). It does this by plotting a straight line through the data points in a way that minimizes the distance between the line and the actual data.The Mathematical FoundationThe model relies on the classic slope-intercept equation you might remember from algebra:$$y = mx + b$$
#### How It Works in Practice: 
#### 1. Training: You feed the algorithm pairs of inputs and outputs.
#### 2. Optimization: The algorithm adjusts the slope and intercept to reduce "loss" (the gap between the predicted line and the real points).
#### 3. Prediction: Once the "best fit" line is established, you can plug in any new value for $x$, and the model will tell you exactly where it falls on that line ($y$).
